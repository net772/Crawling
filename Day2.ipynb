# 한달동안의 영화 데이터 수집 

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd

title_list = []
rate_list = []
rank_list = []
date_list = []

for day in range(20210101,20210119):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=' + str(day)
    # 컴퓨터가 아닌 사람으로 속이는 작업(headers)
    #h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36"}
    #res = req.get(url, headers = h)
    res = req.get(url)
    soup = bs(res.text, 'lxml')
    
    title = soup.select('div.tit5 > a')
    rate = soup.select('.point')
    
    for i in range(len(title)):
        title_list.append(title[i].text)
        rate_list.append(float(rate[i].text))
        rank_list.append(i+1)
        date_list.append(day)
```

```python
len(title_list)
```

```python
dic = {'영화명' : title_list, '평점' : rate_list, '순위' : rank_list, '날짜' :date_list }
```

```python
df = pd.DataFrame(dic)
```

```python
df = df.set_index('순위')
df
```

# 1년동안의 영화데이터 가져오기

```python
date = pd.date_range(start = "2020-01-01", end = "2021-01-18")
date
```

```python
#문자형 포맷팅

days = date.strftime('%Y%m%d')
days
```

# tqdm_notebook >> 진행상황을 알려주는 함수 

```python
from tqdm import tqdm_notebook
```

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd

title_list = []
rate_list = []
rank_list = []
date_list = []

for day in tqdm_notebook(days):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=' + day
    # 컴퓨터가 아닌 사람으로 속이는 작업(headers)
    #h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36"}
    #res = req.get(url, headers = h)
    res = req.get(url)
    soup = bs(res.text, 'lxml')
    
    title = soup.select('div.tit5 > a')
    rate = soup.select('.point')
    
    for i in range(len(title)):
        title_list.append(title[i].text)
        rate_list.append(float(rate[i].text))
        rank_list.append(i+1)
        date_list.append(day)
```

```python
dic = {'영화명' : title_list, '평점' : rate_list, '순위' : rank_list, '날짜' :date_list }
```

```python
df = pd.DataFrame(dic)
df.set_index('순위', inplace = True)
```

```python
df
```

```python
df.to_csv('./movie.csv', encoding = 'euc-kr')
```

# 평점 리뷰

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd

review_list = []
title_list = []

for i in range(1,11):
    url = 'https://movie.naver.com/movie/point/af/list.nhn?&page=' + str(i)
    
    res = req.get(url)
    soup = bs(res.text, 'lxml')
    
    review = soup.select('td.title')
    title = soup.select('td.title>a.color_b')
    
    a = soup.select('td.title > a')
    span = soup.select('td.title > span')
    div = soup.select('td.title > div')
    em = soup.select('td.title > em')

    for i in a:
        i.extract()
    
    
    for i in span:
        i.extract()
    
    for i in div:
        i.extract()

    for i in em:
        i.extract()
        
    for i in range(len(review)):
        review_list.append(review[i].text.strip())
        title_list.append(title[i].text.strip())
```

```python
len(title_list)
```

```python
dic = {'영화제목' : title_list, '리뷰' : review_list}
df = pd.DataFrame(dic)
df
```
