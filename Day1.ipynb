# BeautifulSoup 사용하기
### --- 가져온 페이지 정보에서 내가 원하는 내용만 추출할 때 사용

```python
!pip install beautifulsoup4  
#BeautifulSoup시작 전 설치
```

```python
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.naver.com/'
res = req.get(url)#response[200] = 데이터가 잘 넘어왔어요!
soup = bs(res.text, 'lxml')# bs(어떤걸 가공할건지, 어떻게 가공할건지(파싱방법))
```

```python
soup
```

```python
#find_all = 특정 데이터를 모두 찾아오는 기능(태그, 클래스/ 아이디) -- 잘 사용 안함 select를 주로 사용
result `= soup.find_all('a',class_ = 'nav')
```

```python
result
```

```python
#find_all 과 같은 기능 // 이 방법을 더 많이 사용
result = soup.select('a.nav')
```

```python
result
```

```python
result[2].text
```

```python
for i in result:
    print(i.text)
```
